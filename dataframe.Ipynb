Seleccionar un subconjunto del dataframe 

Filtrar información entree otras cosas 

```python
# Seleccionar un subconjunto para ver los datos que me sirven para hacer diferentes cosas 

llantas.iloc[2:5] # Esta solo trae los primeros 5 elementos del dataframe
```

La diferencia entre un iloc y loc 

iloc -> trae de acuerdo a las posiciones 
loc -> no a la posicción si no al numero que estan en los corchetes o lo valores de los indices 

```python

llantas.loc[2:5]
```

```python
seleccion = llantas[llantas['Empleado'] == 24.0]
seleccion
```

```python
seleccion = llantas.iloc[:, 1:3] # Primero estan las filas y luego las columbas
seleccion
```

```python
seleccion = llantas.loc[:, 'IdCliente':'Empleado']  # Con el loc toca que especificar los indices 
seleccion
```

```python
idCliente = llantas['IdCliente']
idCliente
print(type(idCliente)) # Crea una serie 
```
